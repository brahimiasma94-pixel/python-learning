```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
student={
    "name":["asmaa","sara","yahia","ahmed","lina"],
     "math":[85,90,78,92,70],
     "science":[88,75,95,60,80],
     "enghish":[70,85,80,88,75]
}
df=pd.DataFrame(student)
df
```

```python
df["average"]=np.mean(df[["math","science","enghish"]],axis=1).round(2)
df["status"]=df["average"].apply(lambda x: "pass" if x>=80 else "fail")
df
```

```python
print("===general statistics===")
print (df[["math","science","enghish",	"average"]].describe().round(2))

print("\n=====top student======")
print(df[df["average"]==df["average"].max()])

print("\n=====lowest student===")
print(df[df["average"]==df["average"].min()])
```

```python
colors=["steelblue" if s=="pass" else "salmon" for s in df["status"]]
plt.bar(df["name"],df["average"],color=colors)
plt.title("student average")
plt.xlabel("name")
plt.ylabel("average")
plt.ylim(60,100)
plt.axhline(y=80,color="red",linestyle="--",label="pass line")
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy import stats

grades=np.random.normal(80,10,50).round(2)

stat, p=stats.normaltest(grades)

print("p_value :",round(p,3))

if p > 0.05:
    print("normal distribution")
else:
    print("not normal distribution")
```

```python
plt.figure(figsize=(8,5))
plt.hist(grades,bins=15,color="steelblue",edgecolor="white")
plt.axvline(x=grades.mean(),color="red",linestyle="--",label="Mean")
plt.title("grade distribution")
plt.xlabel("grade")
plt.ylabel("count")
plt.legend()
plt.show()
```
